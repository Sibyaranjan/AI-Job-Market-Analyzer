```python
def convert_experience(exp):
    if pd.isnull(exp):
        return None

    exp = exp.lower().replace('years', '').strip()
    parts = exp.split('to')

    try:
        nums = [int(p.strip()) for p in parts]
        return sum(nums) / len(nums)
    except:
        return None

df['experience_num'] = df['Experience'].apply(convert_experience)  # Convert experience into numeric
```

```python
df['experience_num']
```

```python
def convert_salary(sal):
    if pd.isnull(sal):
        return None

    sal = sal.replace('$', '').replace('K', '')
    parts = sal.split('-')

    try:
        nums = [float(p.strip()) for p in parts]
        return sum(nums) / len(nums)
    except:
        return None

df['salary_num'] = df['Salary Range'].apply(convert_salary)  # Convert salary into numeric column
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['location_encoded'] = le.fit_transform(df['location'])
```

```python
df['skills'] = df['skills'].str.lower()
df['skills'] = df['skills'].str.replace('[^a-zA-Z, ]', '', regex=True)
df['skills'] = df['skills'].str.replace(',', ' ')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=1000)
skills_matrix = tfidf.fit_transform(df['skills'])
```

```python
from scipy.sparse import hstack

numeric_features = df[['experience_num', 'location_encoded']]

X = hstack([skills_matrix, numeric_features])
y = df['salary_num']
```
